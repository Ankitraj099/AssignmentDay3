```python
marks=np.random.normal(70,10,1000)
plt.hist(marks,bins=20)
plt.title("marks distribution")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,5))

Months = ["Jan","Feb","March","April","May","June"]
Sales = [34500,45000,50000,52000,55000,60000]

ax.plot(Months, Sales, color="blue", linewidth=2, marker="o", markersize=7)

ax.set_title("Monthly Sales")
ax.set_xlabel("Months")
ax.set_ylabel("Sales")
ax.grid(axis="y", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import random
heads=0
total=1000
for i in range(total):
    res=random.randint(0,1)
    if(res==1):
        heads+=1
tails=total-heads

print("head", heads)
print ("tail", tails)
print("probability",heads/total)
```

```python
height=np.random.normal(165,8,200)
print("shortest :",round(height.min(),1))
print("tallest :", round(height.max(),1))
print("average :", round(height.mean(),1))
plt.hist(height , bins=20,color="grey",edgecolor="white")

plt.axvline(height.mean(),color="red", linewidth=2,label="Avergae")
plt.title("height of 200 student")
plt.xlabel("height in cm")
plt.ylabel("no of student")
plt.legend()
plt.show()      
      
      
```
